```python
import numpy as np
import matplotlib.pyplot as plt

array = np.random.randint(1, 100, size=(10, 3))


plt.figure(figsize=(10, 6))
plt.imshow(array, cmap='viridis', aspect='auto')
plt.colorbar(label='Value')
plt.title('Viz of Numpy Array with Random Integers')
plt.xlabel('Column')
plt.ylabel('Row')
plt.show()
```
